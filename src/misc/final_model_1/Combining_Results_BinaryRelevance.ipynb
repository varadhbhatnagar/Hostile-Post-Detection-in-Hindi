```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
# Label Mapping
labels_mapping = {'defamation':0,
 'fake':1,
 'hate':2,
 'non-hostile':3,
 'offensive':4}

reverse_labels_mapping = {0:'defamation',
 1:'fake',
 2:'hate',
 3:'non-hostile',
 4:'offensive'}
```

```python
y_pred_defamation = pd.read_csv('y_pred_test_defamation.csv', header=0, index_col='Unique ID')
y_pred_fake = pd.read_csv('y_pred_test_fake.csv', header=0, index_col='Unique ID')
y_pred_hate = pd.read_csv('y_pred_test_hate.csv', header=0, index_col='Unique ID')
y_pred_non_hostile = pd.read_csv('y_pred_test_non_hostile_2.csv', header=0, index_col='Unique ID')
y_pred_offensive = pd.read_csv('y_pred_test_offensive.csv', header=0, index_col='Unique ID')
```

```python
y_pred = y_pred_defamation.join([y_pred_fake, y_pred_hate, y_pred_non_hostile, y_pred_offensive], how='outer')
y_pred.fillna(0, inplace=True)
y_pred.head()
```

```python
data = []
for index, row in y_pred.iterrows():
    st = []
    if row['fake']:
        st.append('fake')
    if row['defamation']:
        st.append('defamation')
    if row['hate']:
        st.append('hate')
    if row['offensive']:
        st.append('offensive')
    if row['non-hostile']:
        st.append('non-hostile')
    if len(st) ==0:
        st.append('offensive')
        st.append('hate')
    data.append(",".join(st))
```

```python
data
```

```python
df = pd.DataFrame(data = data, columns = ["Labels Set"])
```

```python
df.index +=1
df.index.name = 'Unique ID'
```

```python
df.to_csv('submission_testdata_1.csv')
```
